* This script is used to figure out a function to reflect the sound spreading loss. The graphs reveals how much sound decreases as distance grows.

* I assumed that beyond certain point the sound level drops more radically by the distance (after distance > 15 units in the virtual world)

```python
%matplotlib inline
import numpy
import matplotlib.pyplot as plt

e = numpy.exp(1)

r = 1.4 #decides the maximum distance

print "the maximum distance we can hear is %s" %((e+20)/r) # domain of log is x>0

x = numpy.linspace(1, 15, 200)

def f(x):
    return .33*numpy.log(e+20-r*x)

fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)
axes.plot(x, .33*numpy.log(e+20-r*x), 'k--')
axes.set_xlabel("distance from sound")
axes.set_ylabel("volume")
axes.set_title("distance and volume when distance < 15")
plt.show()

print 'when distance is 15, volume is %s' %(f(15))
print 'when distance is 14, volume is %s' %(f(14))
print 'when distance is 10, volume is %s' %(f(10))
print 'when distance is 5, volume is %s' %(f(5))
print 'when distance is 1, volume is %s' %(f(1))
```
